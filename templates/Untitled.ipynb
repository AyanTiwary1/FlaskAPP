```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
url = "https://onlinecourses.swayam2.ac.in/imb23_mg03/unit?unit=11&lesson=12"
urlclient = urlopen(url)
urlread=urlclient.read()
soup=bs(urlread,"html.parser")
print(soup)
bigbox = soup.find_all("div" ,{"id":"pdf-transcript-div"})
print(bigbox)
for i in bigbox:
    image_url =  i['src']
    image_data = requests.get(image_url).content
print(image_data)

```
